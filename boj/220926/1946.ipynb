```python
#input보다 sys가 시간단축됨.

import sys 


case = int(input())

for i in range(case):
    result = []
    person = int(input())

    for i in range (person):
        test1, test2 = map(int, sys.stdin.readline().split())
        result.append([test1,test2]) #리스트에 리스트형태로 넣어줌
    
    result.sort() #test1기준으로 정렬
    rank = result[0][1]     
    count = 1 

    for i in range(1, person):
        if rank > result[i][1]:
            count+=1
            rank = result[i][1]

    print(count)
```
